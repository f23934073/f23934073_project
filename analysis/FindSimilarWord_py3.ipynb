```python
#2GRAM
from operator import itemgetter
def list2freqdict(sentence):#要先算出漢字的頻率分布。我們先定義一個function，input是個list，output是個dictionary：
    chlist=[ch for ch in sentence]#每個字都分開塞入chlist
    mydict_1=dict()
    for ch in chlist:
        if ch == dict:
            mydict_1[ch]=mydict_1.get(ch,0)
        else:
            mydict_1[ch]=mydict_1.get(ch,0)+1
    chbigram =[chlist[i:i+2] for i in range(0,len(chlist)-1)]
    
    bigramfreqdict=dict()
    for (ch1,ch2) in chbigram:
        bigramfreqdict[(ch1,ch2)]=bigramfreqdict.get((ch1,ch2),0)+1
    bigramfreqsorted = sorted(bigramfreqdict.items(), key=itemgetter(1), reverse=True)    
    
    chs=str()
    list1 = []
    for (token,num) in bigramfreqsorted:
        for ch in token:
            chs=chs+ch
        list1.append(chs)
        chs=''
    return list1

#3GRAM
def list3freqdict(sentence):#要先算出漢字的頻率分布。我們先定義一個function，input是個list，output是個dictionary：
    chlist=[ch for ch in sentence]#每個字都分開塞入chlist
    mydict_1=dict()
    for ch in chlist:
        if ch == dict:
            mydict_1[ch]=mydict_1.get(ch,0)
        else:
            mydict_1[ch]=mydict_1.get(ch,0)+1
    chbigram =[chlist[i:i+3] for i in range(0,len(chlist)-2)]
    
    trigramfreqdict=dict()
    for (ch1,ch2,ch3) in chbigram:
        trigramfreqdict[(ch1,ch2,ch3)]=trigramfreqdict.get((ch1,ch2,ch3),0)+1
    trigramfreqsorted = sorted(trigramfreqdict.items(), key=itemgetter(1), reverse=True)    
    
    chs=str()
    list1 = []
    for (token,num) in trigramfreqsorted:
        for ch in token:
            chs=chs+ch
        list1.append(chs)
        chs=''
    return list1

def list4freqdict(sentence):#要先算出漢字的頻率分布。我們先定義一個function，input是個list，output是個dictionary：
    chlist=[ch for ch in sentence]#每個字都分開塞入chlist
    mydict_1=dict()
    for ch in chlist:
        if ch == dict:
            mydict_1[ch]=mydict_1.get(ch,0)
        else:
            mydict_1[ch]=mydict_1.get(ch,0)+1
    chbigram =[chlist[i:i+4] for i in range(0,len(chlist)-3)]

    fourgramfreqdict=dict()
    for (ch1,ch2,ch3,ch4) in chbigram:
        fourgramfreqdict[(ch1,ch2,ch3,ch4)]=fourgramfreqdict.get((ch1,ch2,ch3,ch4),0)+1
    fourgramfreqsorted = sorted(fourgramfreqdict.items(), key=itemgetter(1), reverse=True)    
    chs=str()
    list1 = []
    for (token,num) in fourgramfreqsorted:
        for ch in token:
            chs=chs+ch
        list1.append(chs)
        chs=''
    return list1
```
