```python
import os
import jieba
import jieba.analyse
import re
import pandas as pd 
import simplejson as json
from six import iteritems, itervalues, string_types
from gensim import models
jieba.analyse.set_stop_words("stopwords_1227.txt")
jieba.analyse.set_stop_words("stop.txt")
```

```python
file_array = os.listdir('project5')
contents_cut = []
for j in range(1,len(file_array)):
    path = 'project5/'+file_array[j]
    a = open(path,'r').read().decode('utf-8')
    k = re.sub(u'<br />|；|，|！|[A-Za-z]|,|=|[〕]|\+|[▲]|[＞]|[￣]|[＜]|[☑]|[〔]|[█]|[▓]|$|\{|\}|<|>|「|」|[〃]|[§]|\'|[／]|[↓]|[→]|^|_|@|[＃]|[？]|&|[●]|[～]|3|4|5|6|7|9|。|[.]| |[\n]|、|發信站: 批踢踢實業坊\(ptt.cc\)|（|）|[▼]|[【]|[】]|"|[◆]|※|[[]|[]]|~|/|“|”|[*]|：|%|[?]|-|:|!|…|[★]|·|;|》|[)]|[(]|《|0|　|[1-9]|<\br>|','',a)
    contents_cut.append(jieba.analyse.extract_tags(k,topK=1000))
    
```

```python
f = open('contents_cut1.txt','w')
for i in contents_cut:
    doc = "\n".join(i)
    print doc.encode('utf-8')
    f.write(doc.encode('utf-8'))
f.close()
```

```python
sentences = models.word2vec.LineSentence("contents_cut1.txt")
```

```python
model = models.word2vec.Word2Vec(sentences, size=50, window=1, min_count=1, workers=1)
```

```python
print ' '.join(model.vocab.viewkeys())
```

```python
from collections import Counter
from sklearn import  cluster
import numpy as np
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage
from sklearn.cluster import hierarchical
```

```python
model.init_sims()
```

```python
model.syn0norm
```

```python
kmean = cluster.KMeans(n_clusters=10)
result = kmean.fit(model.syn0norm)
```

```python
Counter(result.labels_)
```

```python
label = 4
print " ".join(np.array(model.index2word)[result.labels_ == label])
```

```python
index = result.labels_ == label
dist = pdist(np.array(model.syn0norm)[index])
m = linkage(dist, method='average')
h = hierarchical.hierarchy.fcluster(m, 0.5*dist.max(), 'distance')
```
