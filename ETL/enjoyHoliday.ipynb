```python
import requests as rs
from bs4 import BeautifulSoup as bs
domain = 'http://www.enjoyholiday.com.tw/list/%E5%8F%B0%E5%8C%97/c/'
res = rs.get('http://www.enjoyholiday.com.tw/list/%E5%8F%B0%E5%8C%97/c/1/')
res.encoding = 'utf-8'
soup = bs(res.text)
fid = open("python/enjoyHoliday.txt","w")
for j in range(2,12):
    for body in soup.select('tr'):
        if body.select('strong a'):
            a = str(body.select('strong a')[0].text.encode('utf-8'))
            b = str(body.select('span')[0].text.encode('utf-8'))
            txt = a + "," + b
            print txt
            fid.write(txt)
            fid.write('\n')
        res = rs.get(domain + str(j)+"/")
        res.encoding = 'utf-8'
        soup = bs(res.text)
fid.close()
```
